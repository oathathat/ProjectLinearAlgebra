```python
import pandas as pd
import requests

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import warnings
warnings.filterwarnings('ignore')

datalist = []
for i in range (201,394):
    url= f'https://covid19.ddc.moph.go.th/api/Cases/round-3-line-lists?page={i}'
    r = requests.get(url)
    j = r.json()
    temp = j['data'].copy()
    datalist.extend(temp)
df = pd.DataFrame(datalist)    
df.to_csv('updateCovid19.csv', index=False)
```
